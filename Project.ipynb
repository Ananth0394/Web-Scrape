```python
def scrape():
    for link in urllist:
        t1=time.time()
        chrome_options = Options()
        chrome_options.add_argument('--headless')
        chrome_options.add_argument('--no-sandbox')
        chrome_options.add_argument('--disable-dev-shm-usage')
        driver = webdriver.Chrome( options = chrome_options)
        driver.get(link)
        html = driver.page_source # Get the html of the page
        driver.quit() # Close the browser
        title=html[html.find('h1 class="b_42 company_name"')+29:html.find('</h1')]
        complist.append(title)
        temphtml=html[html.find('h1 class="b_42 company_name"')+29:]
        temphtml=temphtml[temphtml.find('<div class="PB10"'):]
        #html=html[html.find('<div class="FL gry10"'):]
        soup = bs(temphtml, 'html.parser')
        temphtml=temphtml[temphtml.find("NSE")+5:]
        NSE=temphtml[:temphtml.find("<")]
        NSElist.append(NSE)
        soup = bs(html, 'html.parser')
        temp=soup.findAll("div",{"class":"brdb PB5"})
        ctr=0
        for k in temp:
            if(len(k.findAll("div",{"class":"FR MT2 live"}))):
                #print(k.get_text())
                if(ctr==0):
                    stock=k.get_text()
                    stocklist=stock.splitlines(0)
                    #print(stocklist[7].split(" "))
                    bseprice_after.append(stocklist[6])
                    bsestock_percentchange.append(float(stocklist[7].lstrip().split(" ")[1][1:-2]))
                    bsestock_pricechange.append(float(stocklist[7].lstrip().split(" ")[0]))
                    ctr+=1
                else:
                    stock=k.get_text()
                    stocklist=stock.splitlines(0)
                    nseprice_after.append(stocklist[6])
                    nsestock_percentchange.append(float(stocklist[7].lstrip().split(" ")[1][1:-2]))
                    nsestock_pricechange.append(float(stocklist[7].lstrip().split(" ")[0]))
                    ctr+=1
        t2=time.time()
        print("Time:",t2-t1)
        t=random.randint(3,5)
        time.sleep(t)
```
